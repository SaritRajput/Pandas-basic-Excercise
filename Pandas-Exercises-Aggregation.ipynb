Use numpy.where() to encode gender column.

```python
user=np.where(usr.gender=='M',)
```

```python
usr=pd.crosstab(user.gender,user.occupation,margins=True)

usr
```

```python
male_ratio = user.pivot_table(index='occupation', columns='gender', aggfunc='size', fill_value=0)
```

```python
sums = male_ratio[['F', 'M']].sum(axis=1)
male_ratio['MaleRatio'] = round(100 * male_ratio['M'] / sums , 1)

# result
male_ratio['MaleRatio'].sort_values(ascending=False)
```

### Step 8.  For each occupation present the percentage of women and men

```python
M_F_ratio = user.pivot_table(index='occupation', columns='gender', aggfunc='size', fill_value=0)
```

```python
sums = M_F_ratio[['F', 'M']].sum(axis=1)
M_F_ratio['MaleRatio'] = round(100 * M_F_ratio['M'] / sums , 1)
M_F_ratio['FemaleRatio'] = round(100 * M_F_ratio['F'] / sums , 1)

# result
print(M_F_ratio)
```
